## tariffs

Congressional statements on tariffs express broad concern that sweeping or rapidly shifting tariff policies—especially those targeting China—impose significant costs on American consumers, farmers, and manufacturers. Lawmakers repeatedly note that China’s retaliatory tariffs and supply-chain disruptions have harmed U.S. exporters and increased economic uncertainty. Overall, the sentiment is that tariffs should be narrowly targeted at unfair Chinese trade practices rather than used indiscriminately in ways that weaken U.S. competitiveness and inadvertently strengthen China’s global position.

## Russia or Russia-China

Lawmakers consistently portray Russia and China as central, tightly linked adversaries—often part of a broader “axis” that includes Iran and North Korea—seeking to roll back U.S. power, undermine democracy, and expand authoritarian influence worldwide. They warn that when the United States retreats from global leadership, cuts foreign aid, weakens institutions like USAID, or creates internal chaos, Russia and China quickly move to fill the vacuum through disinformation, economic coercion, military buildup, cyber intrusions, and soft-power competition. Across the statements, there is a clear call for a whole-of-government strategy that both counters growing Russia–China cooperation (including in nuclear arms, propaganda, and regional influence) and reinforces U.S. alliances, values, and global credibility.

## Chinese government

The excerpts depict the Chinese Government as a powerful, authoritarian actor aggressively seeking to expand its global influence through technology acquisition, cyber espionage, political interference, and strategic investments. U.S. lawmakers express deep concern that Beijing’s actions—ranging from hacking telecommunications networks to leveraging platforms like TikTok to influence Americans—pose escalating national security, economic, and geopolitical threats. Across the statements, there is strong bipartisan urgency for the United States to harden protections, prevent sensitive technologies from reaching China, and counter the CCP’s expanding reach into international institutions, global markets, and critical infrastructure

## nucleaer

The excerpts describe rising fears of a renewed nuclear arms race among the United States, China, and Russia, driven by China’s rapid arsenal expansion, Russia’s signals toward renewed testing, and political pressure in the U.S. to build more weapons. Lawmakers argue that weakening arms control agreements and expanding nuclear programs heighten the risk of miscalculation, proliferation, and catastrophic conflict. The statements emphasize an urgent need to restore and strengthen nuclear arms control frameworks to stabilize global security.

## Ukraine

Ukraine is repeatedly framed as the frontline of a global struggle between democracies and authoritarian powers, with China and Russia closely watching how the United States responds. Lawmakers warn that abandoning or hesitating to support Ukraine would embolden adversaries like Xi Jinping and Vladimir Putin, increase the risk to Taiwan, and undermine U.S. credibility with allies. The excerpts emphasize that the outcome of the war in Ukraine has direct implications for global security, great-power competition, and America’s ability to deter further aggression by China, Russia, Iran, and North Korea.

## percent

Many statements reference specific percentage values to highlight the scale of tariffs, trade measures, and global production shares involving China, Canada, Mexico, and other countries. These percentages are used to demonstrate how deeply intertwined supply chains are, how tariffs dramatically raise costs, and how dominant China is in sectors such as batteries, solar panels, EVs, steel, and global manufacturing. Overall, the repeated use of percentage figures underscores the magnitude of U.S.–China economic tensions and the potential economic consequences for American industries, consumers, and strategic competitiveness.

## step

China repeatedly “steps in” to fill strategic, economic, and diplomatic vacuums whenever the United States withdraws from global leadership or cuts foreign aid programs. Lawmakers also describe specific “steps” the U.S. must take—such as tightening financial oversight, derisking supply chains, and strengthening alliances—to counter China’s growing influence. Overall, “step” is used to highlight both the dangers of U.S. retreat and the urgent actions required to prevent China from gaining further geopolitical advantage.

## ties

The term “ties” is used to describe both positive and negative relationships: on the positive side, it highlights efforts to deepen economic, tax, security, and travel cooperation between the United States and Taiwan, as well as long-standing people-to-people and educational links. On the negative side, it repeatedly flags concerning ties between Chinese companies, the Chinese Communist Party, and various foreign firms, universities, or technology partners that may create national security, human rights, or financial risks for the United States. Overall, “ties” captures the tension between building trusted partnerships (especially with Taiwan and allies) and scrutinizing or severing problematic connections linked to the PRC and CCP.

## going

U.S. lawmakers repeatedly warn that if Washington “steps back,” China is going to fill the strategic, economic, technological, and geopolitical void. Across debates—from trade and AI to military posture and global development—Members emphasize that China is going to expand influence wherever the U.S. hesitates, undermining American competitiveness and security. The term highlights a forward-looking concern: decisions made now are going to determine whether the United States or China shapes the next era of global leadership.

## USAID

Lawmakers repeatedly argue that shutting down or weakening USAID directly benefits China and Russia, because USAID is one of America’s main tools for countering their global influence through development, stabilization, and soft-power programs. Multiple statements warn that eliminating USAID would cede leadership, allow China to dominate strategic regions through Belt and Road–style initiatives, and leave fragile countries vulnerable to authoritarian influence, misinformation, and extremist groups. Overall, USAID is portrayed as essential to U.S. national security, global competitiveness, and the ability to shape international norms in opposition to China’s expanding reach.

## cede

References to “ceding” emphasize a fear that U.S. policies—particularly tariffs, energy restrictions, or weakened investments—could surrender economic and technological leadership to China. Lawmakers argue that failing to invest in renewable energy, biomedical research, or advanced technologies would allow China to dominate global markets, take American jobs, and gain strategic advantages. Overall, “cede” is used to warn that the United States must not relinquish its competitive edge or geopolitical influence to China.

## North Korea

Mentions of North Korea consistently place it within a hostile bloc alongside China, Russia, and Iran, portraying these states as an interconnected axis challenging U.S. interests and global stability. North Korea is cited both as a direct adversary—through cyber threats, military cooperation with Russia, and regional aggression—and as a partner enabled or influenced by China, reinforcing concerns about authoritarian alignment. Overall, the references frame North Korea as a key component of a broader authoritarian coalition whose actions require strong U.S. deterrence, intelligence capabilities, and steadfast support for allies like Taiwan and Ukraine.

## Elon Musk

References to Elon Musk consistently portray him as deeply entangled with China due to his extensive business operations, competitive pressures, and financial dependencies there. Lawmakers express concern that Musk’s actions—such as allegedly shutting down USAID, influencing foreign aid policy, or gaining access to federal systems—may inadvertently benefit China or align with Beijing’s interests, raising national security and conflict-of-interest questions. Overall, Musk is framed as an unelected, unaccountable actor whose global business incentives could compromise U.S. foreign policy, cybersecurity, and democratic institutions.

## Donald Trump

References to Donald Trump consistently portray him as ineffective or contradictory in confronting China—often taking actions that inadvertently strengthen Beijing’s leverage, expand its global influence, or undermine U.S. competitiveness. His tariff policies are described as isolating the United States rather than China, pushing allies toward Beijing, harming American industries, and enabling China to gain markets, supply chains, and technological advantages. Lawmakers also criticize Trump for reversing export restrictions, enabling advanced chip sales to China, and repeatedly “caving” to Xi Jinping, framing his China policy as erratic, self-defeating, and strategically damaging to U.S. interests

## abroad

Mentions of “abroad” emphasize how U.S. global engagement is essential in an era of rising competition with China, highlighting that American influence, diplomacy, and security efforts overseas are critical to countering Beijing’s expanding reach. Lawmakers warn that underinvesting in foreign assistance, military readiness, and diplomatic presence abroad creates vulnerabilities—allowing China to shape global narratives, gain economic advantages, and deepen its footprint in strategic regions. The term also appears in contexts where U.S. jobs, investments, and policies have been affected by global dynamics, particularly China’s economic actions and aggression abroad.

## did

References to “did” emphasize how China’s actions—such as predatory lending, rapid technological advancement, manufacturing expansion, and suppression of freedoms in Hong Kong—have reshaped geopolitical and economic dynamics. Lawmakers repeatedly contrast what China did with what the United States failed to do, warning that American policy missteps, including tariff strategies and withdrawal from global engagement, have opened strategic gaps that China has eagerly filled. The term also appears in critiques of political decisions, highlighting how U.S. leaders’ actions—or inactions—either strengthened China’s position or weakened America’s alliances and global leadership

## said

References to “said” highlight how U.S. lawmakers repeatedly quote statements made by political leaders, experts, and foreign officials to underscore rising strategic competition with China and the consequences of weakening U.S. institutions like USAID. Many of these quoted remarks warn that China is expanding its influence—economically, militarily, and diplomatically—while U.S. actions such as tariff volatility, foreign-aid cuts, and inconsistent leadership risk ceding global leadership. The term also appears in criticisms of former President Trump’s comments about China, Xi Jinping, and tariff policy, illustrating concerns that rhetoric and policy decisions are empowering authoritarian rivals instead of strengthening American alliances and competitiveness

## Chinese American

The excerpts highlight positive contributions of Chinese American individuals and organizations in the United States. They describe community celebrations such as the Lunar New Year event in Michigan, recognize Chinese American civic groups, and honor notable Chinese Americans—including Kent Wong, a fifth-generation Chinese American dedicated to public service, and others active in local associations or military service. ## clean

References to clean energy emphasize that the United States risks losing global leadership in renewable and low-carbon technologies to China, which is investing heavily—up to $1 trillion a year—in this sector. Lawmakers warn that abandoning or weakening U.S. clean-energy initiatives would cede jobs, manufacturing capacity, and strategic competitiveness to China, Japan, South Korea, and other rivals. Overall, the passages frame clean energy as both an economic necessity and a geopolitical race in which China is rapidly pulling ahead.

## leadership

The passages describe a recurring theme: U.S. global leadership is eroding, creating a vacuum that China—and sometimes Russia—quickly fills. Weakening agencies like USAID, abandoning clean-energy competitiveness, and retreating from international commitments are portrayed as actions that hand diplomatic, economic, technological, and strategic leadership to the People’s Republic of China. Speakers argue that sustained American leadership is essential to uphold the post-WWII international order, counter authoritarian influence, and prevent China from dominating key sectors such as development, energy, technology, and global governance
